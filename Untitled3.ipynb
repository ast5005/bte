```julia
using constants
using types
using Optim
using PyPlot
using data
include("DOSfunctions.jl")
include("userfunctions.jl")
include("propertyfuncs.jl")
```

```julia
Es=collect(0.001:0.001:0.2)
Ec(var)=0.0
effMassC(var)=1.0
var0=[300.0,0.3]
alphaA=collect(0.0:0.2:1.0)
bandC=parBandTx(effMassC(var0),0.0,Ec(var0),[Ec],[effMassC],var0)
figure()
dos=Array{Float64,2}(length(alphaA),length(Es))
for i in 1:length(alphaA)
    bandC=parBandTx(effMassC(var0),alphaA[i],Ec(var0),[Ec],[effMassC],var0)
    dos[i,:]=getDOS_SingleBand_E(bandC,Es)
    labely="alpha",alphaA[i]
    plot(Es,dos[i,:],label=labely)
end
legend(loc=2)
grid()
```
