```julia
parBand(1.0*me,0.0,0.0)
```

```julia
E=collect(0.0:0.01:1.0)
```

```julia
dos=getDOS_E(parBand(1.0*me,0.0,0.0),0.0,1.0,0.01)
```

```julia
using PyPlot
```

```julia
xlim(-0.5,1.5)
xticks(collect(-0.5:0.2:1.5))
ylim(0.0e46,5.0e46)
yticks(collect(0.0e46:0.5e46:5.0e46))
xlabel("Energy (eV)")
ylabel("Density of States (1/Jm3)")
plot(E,dos,linestyle="dashed",marker="o",color="green")
title("Density of States for meff=1.0xme")
```

```julia
doshigheffmass=getDOS_E(parBand(3.0*me,0.0,0.0),0.0,1.0,0.01)
```

```julia
dosloweffmass=getDOS_E(parBand(0.1*me,0.0,0.0),0.0,1.0,0.01)
```

```julia
xlim(-0.5,1.5)
xticks(collect(-0.5:0.2:1.5))
ylim(0.0e46,5.0e46)
yticks(collect(0.0e46:0.5e46:5.0e46))
xlabel("Energy (eV)")
ylabel("Density of States (1/Jm3)")
plot(E,dos,color="blue")
title("Density of States for meff=1.0xme")
```

```julia
using PyPlot
using PyCall
xlim(-0.5,1.5)
xticks(collect(-0.5:0.2:1.5))
ylim(0.0e46,15.0e46)
yticks(collect(0.0e46:2e46:14.0e46))
xlabel("Energy (eV)")
ylabel("Density of States (1/Jm3)")
plot(E,dos,linestyle="dashed",marker="o",color="green")
plot(E,doshigheffmass,linestyle="dashed",marker="o",color="blue")
plot(E,dosloweffmass,linestyle="dashed",marker="o",color="red")
#fig=figure()
#ax=fig[:add_subplot](1,1,1)
#ax.tick_params(axis='x',which='minor',bottom='off')
```
